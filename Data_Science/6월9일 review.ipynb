#### **List Comprehension 이용하기**

```python
lst = [i for i in range(10)]
# i를 출력하라 / for i in range(10)을 돌려서/
lst
```

```python
lst2 = [i ** 2 for i in range(10)]
# i ** 2을 출력하라 / for i in range(10)을 돌려서/
lst2
```

```python
lst3 = [i for i in range(10) if i >= 5]
# i를 출력하라 / for i in range(10)을 돌려서 / if i가 5보다 크거나 같다면
lst3
```

```python
# 1. 리스트 준비하기
seq = ['alpha', 'bravo', 'charlie', 'delta']

# 2. 딕셔너리 만들기 -> 인덱스 정보를 key로, data를 value로 만듦
dict(enumerate(seq))

# enumerate()함수는 벨류에 자동으로 0번부터 인덱스를 넣어준다.
```

### 리스트 딕셔너리 만들기 - 키 순서열과 값 순서열에 클래스 생성자인 zip(key, value) 사용

```python
kseq = 'abcd'
# 문자열 또한 순서열이다.
vseq = ['alpha', 'bravo', 'charlie', 'delta']
# value에 넣을 값들을 리스트로 만듦
dict(zip(kseq, vseq))
# zip()함수를 사용하여 kseq를 key로 vseq를 value로 만들고 dict()함수로 딕셔너리화 했다.
```

### Counter로 세기
* 딕셔너리에서 다른 컬렉션 안의 아이템 갯수를 셀 때 사용한다.

```python
from collections import Counter

phrase = 'a man a plan a canal a panama'

Counter(phrase.split())
# a는 3번, canal은 1번, man은 1번, panama는 1번, plan은 1번

cntr = Counter(phrase.split())

cntr.most_common(1)
# 1번째로 가장 많이 등장한 문자열 세기
```

#텍스트 다루기
* BeautifulSoup을 이용하여 HTML 파싱 (필요한 정보만 가져옴)

```python
from bs4 import BeautifulSoup
from urllib.request import urlopen

# 1. 문자열에서 soup 생성하기
soup1 = BeautifulSoup("<HTML><HEAD><<headers>></HEAD><<body>></HTML>")

# 2. 로컬 파일에서 soup을 생성한다.
soup2 = BeautifulSoup(open('myDoc.html'))

# 3. 웹 문서에서 soup을 생성한다. - 크롤링의 기초단계
soup3 = BeautifulSoup(urlopen('http://www.networksciencelab.com/'))
```

```python
# 'http://www.networksciencelab.com/' 사이트에서 하이퍼 링크 정보 모두 가져오기

with urlopen('http://www.networksciencelab.com/') as doc:
  soup = BeautifulSoup(doc)
#soup # 열고자하는 url의 HTML 구조를 보여준다.

# with urlopen('url주소') as 별칭:
#   soup = BeautifulSoup(doc)

#읽어들인 url의 HTML에서 원하는 원하는 정보(하이퍼링크 정보)만 찾아보기
#하이퍼링크만 추출하기 (list comprehension으로 한 줄로 끝냄)
links = [(link.string, link['href']) for link in soup.find_all('a') if link.has_attr('href')]
# link의 string과 href(Hypertext Refenence)를 추출하라 
# / soup.find_all('a')은 a(anchor)의 모든 값을 찾아 link에 넣는 for 반복문을 수행해라
# / 만약에 link가 href 속성을 갖는다면 (하이퍼링크를 갖는다면) 

links
```
