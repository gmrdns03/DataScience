```python
# 최고기온의 값이 8월에 해당하는 값만 추출해서 그래프로 시각화

import csv
import matplotlib.pyplot as plt



f = open('/content/Suwon.csv', 'r', encoding='cp949')
data = csv.reader(f)
header = next(data)

result = [] # 최고 기온 데이터 중 8월 데이터만 저장할 리슽트를 생성하기

for row in data:
  if row[-1] != '':   # 최고기온 결측치 제외
    if row[0].split('-')[1] == '08':    # 날짜를 잘라서 08월 데이터만 추출하기
      result.append(float(row[-1]))     # 결과 리스트에 담기

plt.figure(dpi=200) # dpi는 해상도를 의미한다. // # figsize는 가로 세로 인치를 구할 수 있다.
plt.plot(result, 'deeppink') # 최고기온 데이터를 빨간색으로 그림
plt.title('August Highest temp of Suwon')
plt.show()

```

```python
# 최고기온의 값이 8월에 해당하는 값만 추출해서 그래프로 시각화

import csv
import matplotlib.pyplot as plt



f = open('/content/Suwon.csv', 'r', encoding='cp949')
data = csv.reader(f)
header = next(data)

result = [] # 최고 기온 데이터 중 8월 데이터만 저장할 리슽트를 생성하기

for row in data:
  if row[-1] != '':   # 최고기온 결측치 제외
    if row[0].split('-')[1] == '06' and row[0].split('-')[2] == '09':    # 매년 6월 9일 데이터를 추출
      result.append(float(row[-1]))     # 결과 리스트에 담기

plt.figure(dpi=200) # dpi는 해상도를 의미한다. // # figsize는 가로 세로 인치를 구할 수 있다.
plt.plot(result, 'deeppink') # 최고기온 데이터를 빨간색으로 그림
plt.show()

```

매년 생일 기준 최고 기온, 최저기온, 평균 기온 그래프 그리기 

```python
import csv
import matplotlib.pyplot as plt
import numpy as np


f = open('/content/Suwon.csv', 'r', encoding='cp949')
data = csv.reader(f)
header = next(data)

high = []
low = []
avg = []

for row in data:
  if row[-1] != '' and row[-2] != '' and row[-3]:   # 최고, 최저, 평균 결측치 제거
    if 1994 <= int(row[0].split('-')[0]):   # 문자열과 비교할 때 같은 것은 비교가능하지만 크고 작은 것은 비교가 불가능하다.
      if row[0].split('-')[1] == '12' and row[0].split('-')[2] == '20':
        high.append(float(row[-1]))
        low.append(float(row[-2]))
        avg.append(float(row[-3]))




plt.figure(dpi=200) # dpi는 해상도를 의미한다. // # figsize는 가로 세로 인치를 구할 수 있다.
plt.rc('font', family = 'NanumBarunGothic') # 나눔 바른 고딕을 기본 글꼴로 설정
plt.rcParams['axes.unicode_minus'] = False # 마이너스 기호 깨짐 방지
plt.plot(high, 'deeppink', label='highest temp') # 최고기온
plt.plot(low, 'skyblue', label='lowest temp') # 최저기온
plt.plot(avg, 'y', label='average temp') # 평균기온
plt.title('내 생일')
plt.legend()
plt.show()
```
