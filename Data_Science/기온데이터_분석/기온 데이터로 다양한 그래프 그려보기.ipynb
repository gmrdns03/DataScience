최고 기온 데이터 히스토그램으로 표현하기

```python

import csv
import matplotlib.pyplot as plt



f = open('/content/Suwon.csv', 'r', encoding='cp949')
data = csv.reader(f)
header = next(data)

result = []

for row in data:
  if row[-1] != '':
    result.append(float(row[-1]))

# 최고기온 데이터 히슽토그램


plt.figure(dpi=200)
plt.hist(result, bins=1000, color='r') # 온도 분포를 셀 수 있다. // 온도 범위를 확인할 수 있다. // 아예 단순하거나 아예 복잡하게 만들기
plt.title('Highest temp of Suwon')
plt.show()
```

8월의 최고 기온 데이터 히스토그램으로 표현하기

```python
import csv
import matplotlib.pyplot as plt



f = open('/content/Suwon.csv', 'r', encoding='cp949')
data = csv.reader(f)
header = next(data)

aug = [] # 8월의 최고 기온을 담을 aug 리스트 변수 선언

for row in data:
  month = row[0].split('-')[1]
  
  if row[-1] != '':
    if month == '01':
      aug.append(float(row[-1]))

# 최고 기온 데이터 히스토그램
plt.figure(dpi=200)
plt.hist(aug, bins=1000, color='r') # 온도 분포(범위)를 셀 수 있다. // 아예 단순하거나 아예 복잡하게 만들기
plt.title('August temp of Suwon')
plt.show()
```

1월과 8월의 데이터 히스토그램으로 시각화하기

```python
import csv
import matplotlib.pyplot as plt



f = open('/content/Suwon.csv', 'r', encoding='cp949')
data = csv.reader(f)
header = next(data)

jan = [] # 1월의 최고 기온을 담을 jan 리스트 변수 선언
aug = [] # 8월의 최고 기온을 담을 aug 리스트 변수 선언

for row in data:
  month = row[0].split('-')[1]
  
  if row[-1] != '':
    if month == '01':
      jan.append(float(row[-1])) # 1월 데이터면 jan 변수에 담기
    if month == '08':
      aug.append(float(row[-1])) # 8월 데이터면 aug 변수에 담기

# 최고 기온 데이터 히스토그램
plt.figure(dpi=200)
plt.hist(jan, bins=100, color='b', label='Aug') # 온도 분포(범위)를 셀 수 있다. // 아예 단순하거나 아예 복잡하게 만들기
plt.hist(aug, bins=100, color='r', label='Jan')
plt.title('January & August temp of Suwon')
plt.show()
```

1월, 6월, 8월 데이터를 히스토그램으로 시각화하기

```python
import csv
import matplotlib.pyplot as plt



f = open('/content/Suwon.csv', 'r', encoding='cp949')
data = csv.reader(f)
header = next(data)

jan = [] # 1월의 최고 기온을 담을 jan 리스트 변수 선언
aug = [] # 8월의 최고 기온을 담을 aug 리스트 변수 선언
apr = [] # 6월의 최고 기온을 담을 aug 리스트 변수 선언

for row in data:
  month = row[0].split('-')[1]
  
  if row[-1] != '':
    if month == '01':
      jan.append(float(row[-1])) # 1월 데이터면 jan 변수에 담기
    if month == '08':
      aug.append(float(row[-1])) # 8월 데이터면 aug 변수에 담기
    if month == '04':
      apr.append(float(row[-1]))




# 최고 기온 데이터 히스토그램
plt.figure(dpi=200)
plt.hist(jan, bins=100, color='b', label='Aug') # 온도 분포(범위)를 셀 수 있다. // 아예 단순하거나 아예 복잡하게 만들기
plt.hist(aug, bins=100, color='r', label='Jan')
plt.hist(apr, bins=100, color='y', label='Apr')
plt.title('January & April & August temp of Suwon')
plt.legend()
plt.show()
```

1월부터 12월까지 최고기온 데이터 상자 그림으로 표현하기

```python
import csv
import matplotlib.pyplot as plt



f = open('/content/Suwon.csv', 'r', encoding='cp949')
data = csv.reader(f)
header = next(data)

month = [[], [], [], [], [], [], [], [], [], [], [], []]
# 월 별 데이터 저장할 month 변수 선언


for row in data:
  if row[-1] != '' :
    month[int(row[0].split('-')[1]) -1].append(float(row[-1]))
    # 월과 같은 번호의 인덱스에 월별 데이터 저장
    # 1월 row[0].split('-')[1] -> month[0]



plt.figure(dpi=200)
plt.boxplot(month)
plt.show()


```

8월 1일부터 31일까지 최고기온 데이터 상자 그림으로 표현하기

```python
import csv
import matplotlib.pyplot as plt



f = open('/content/Suwon.csv', 'r', encoding='cp949')
data = csv.reader(f)
header = next(data)

day = []
# 일별 데이터를 저장할 리스트 day 변수 선언

for i in range(31):
  day.append([])  # day 리스트내에 31개의 리스트를 생성
for row in data:
  if row[-1] != '' :
    if row[0].split('-')[1] == '08':
      day[int(row[0].split('-')[2])-1].append(float(row[-1]))

plt.figure(dpi=200)
plt.boxplot(day)
plt.show()
```
