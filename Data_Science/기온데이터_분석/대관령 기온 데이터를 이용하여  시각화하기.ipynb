```python
# 1) 히스토그램을 활용한 최고기온, 최저 기온, 평균기온 분포 확인하기

import csv
import matplotlib.pyplot as plt


f = open('/content/Deagwanryeong.csv', 'r', encoding='cp949')
data = csv.reader(f)
header = next(data)

high = []


for row in data:
  if row[-1] != '':
    high.append(float(row[-1]))

plt.figure(dpi=200)
plt.title('Highest temp of Deagwanryeong')
plt.hist(high, bins = 100, color='r')
plt.show()

f.close()
```

```python
# 1) 히스토그램을 활용한 최고기온, 최저 기온, 평균기온 분포 확인하기

import csv
import matplotlib.pyplot as plt


f = open('/content/Deagwanryeong.csv', 'r', encoding='cp949')
data = csv.reader(f)
header = next(data)

low = []


for row in data:
  if row[-1] != '':
    low.append(float(row[-2]))

plt.figure(dpi=200)
plt.title('Lowest temp of Deagwanryeong')
plt.hist(low, bins = 100, color='b')
plt.show()

f.close()
```

```python
# 1) 히스토그램을 활용한 최고기온, 최저 기온, 평균기온 분포 확인하기

import csv
import matplotlib.pyplot as plt


f = open('/content/Deagwanryeong.csv', 'r', encoding='cp949')
data = csv.reader(f)
header = next(data)

high = []
low = []
avg = []

for row in data:
  if row[-1] != '' and row[-2] != '' and row[-3] != '':
    high.append(float(row[-1]))
    low.append(float(row[-2]))
    avg.append(float(row[-3]))

plt.figure(dpi=200)
plt.title('Highest & Lowest & Average temp of Deagwanryeong')
plt.hist(high, bins = 100, color='r')
plt.hist(low, bins = 100, color='b')
plt.hist(avg, bins = 100, color='y')
plt.show()

f.close()
```

```python
#2) 상자그림을 활용한 최고 기온 vs 서울 최고 기온

import csv
import matplotlib.pyplot as plt

f_1 = open('/content/seoul.csv', 'r', encoding='cp949')
data1 = csv.reader(f_1)
header1 = next(data1)


f = open('/content/Deagwanryeong.csv', 'r', encoding='cp949')
data = csv.reader(f)
header = next(data)

high = []
seoul_high = []

for row_1 in data1:
  if row_1[-1] != '':
    seoul_high.append(float(row_1[-1]))

for row in data:
  if row[-1] != '':
    high.append(float(row[-1]))

plt.figure(dpi=200)
plt.title('Highest & Lowest temp of Deagwanryeong')
plt.boxplot([high, seoul_high])
plt.show()
f.close()
```

```python
# 3) 월별 최고기온 상자 그림 그리기

import csv
import matplotlib.pyplot as plt

f = open('/content/Deagwanryeong.csv', 'r', encoding='cp949')
data = csv.reader(f)
header = next(data)

month = []

for i in range(12):
  month.append([])

for row in data:
  if row[-1] != '':
    month[int(row[0].split('-')[1]) -1].append(float(row[-1]))

plt.figure(dpi=200)
plt.boxplot(month, showfliers = False)
# showfliers = False 는 아웃라이어들을 제거해준다.
plt.title('Month temp of Deagwanryeong')
plt.show()


```

```python
# 4) 특정 달의 일별 최고 기온 상자 그림 그리기

import csv
import matplotlib.pyplot as plt

f = open('/content/Deagwanryeong.csv', 'r', encoding='cp949')
data = csv.reader(f)
header = next(data)

day = []

for i in range(31):
  day.append([])

for row in data:
  if row[-1] != '':
    if row[0].split('-')[1] == '01':
      day[int(row[0].split('-')[2])-1].append(float(row[-1]))

plt.figure(figsize=(10,2), dpi=200)
plt.boxplot(day, showfliers = False)
plt.title('January temp of Deagwanryeong')
plt.show()

```
