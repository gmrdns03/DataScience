##matplotlib을 이용한 기본 그래프 그리기

내가 갖는 데이터를 어떻게 잘 표현할 수 있을까 생각을 하고 차트를 고르기

import matplotlib.pyplot as plt

plt.plot(데이터 셋) 그래프 그리기

```python
import matplotlib.pyplot as plt
# 라이브러리 import

plt.plot([10, 20, 30, 40]) # Y값이 주어지면 자동으로 x값을 0부터 1씩 증가한다!!!
plt.show
# Y값이 가장 중요함!! Y는 정답임!!
```

plt.plot(x, y)그래프 그리기

```python
import matplotlib.pyplot as plt


plt.plot([1, 2, 3, 4], [15, 11, 37, 46]) 
# plt.plot(x, y) 함수에 데이터 입력하기
plt.show
# 그래프 보여주기
```

```python
import matplotlib.pyplot as plt

plt.title('plotting') # 타이틀 속성 추가(한글은 깨지므로 영어로 하기)
plt.plot([1, 2, 3, 4], [15, 11, 37, 46]) 

plt.show
```

```python
import matplotlib.pyplot as plt

plt.title('plotting')

plt.plot([10, 20, 30, 40], label='asc') # 증가를 의미하는 asc 범례

plt.plot([40, 30, 20, 10], label='desc') # 감소를 의미하는 desc 범례


plt.legend(loc=9) 
# 그래프에 범례를 넣기 위해 꼭 사용!!
# loc= 숫자 를 사용해서 범례의 위치를 바꿔줄 수 있다.
# 차트의 선을 절대로 범례가 침범하면 안된다!!
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.title('color')

plt.plot([10, 20, 30, 40], color='R' , label='RED') # 기본 색상은 약자로 가능하다.(R, G, B, K(BLACK), Y)

plt.plot([40, 30, 20, 10], color='G' ,label='GREEN') # 핑크색 그래프


plt.legend() 

plt.show()
```

```python
import matplotlib.pyplot as plt

plt.title('Line Style')

plt.plot([10, 20, 30, 40], color='R' , linestyle = '--',label='Dashed') 
# 빨간색 dashed 그래프
plt.plot([40, 30, 20, 10], color='G', ls=':' ,label='GREEN')
# 초록색 dotted 그래프linestyle -> ls로 표현 가능

plt.legend() 

plt.show()
```

```python
import matplotlib.pyplot as plt

plt.title('Marker')

plt.plot([10, 20, 30, 40], 'r.', label='circle') 
# 빨간색 dashed 그래프
plt.plot([40, 30, 20, 10], 'g^',label='triangle up')
# 초록색 dotted 그래프linestyle -> ls로 표현 가능

plt.legend() 

plt.show()
```

```python
import matplotlib.pyplot as plt

plt.title('Marker')

# <색상> <마커모양> <선모양> 코드로 작성가능

plt.plot([10, 20, 30, 40], 'r.-', label='circle') 
# 빨간색 circle 그래프
plt.plot([40, 30, 20, 10], 'g^--',label='triangle up')
# 초록색 triangle up 그래프 

plt.legend() 
plt.show()
```

데이터 시각화하기

```python
import csv
import matplotlib.pyplot as plt

f = open('/content/seoul.csv', 'r', encoding='cp949')

data = csv.reader(f)
header = next(data)

result = []
# 결측치를 제거한 최고기온 리스트

for row in data:
  if row[-1] != '':
    # 최고기온이 ''(빈 문자열)이 아니면, 
    result.append(float(row[-1])) 
    # 실수형으로 바꾸고 리스트에 넣기

plt.figure(dpi=200)
plt.plot(result, 'r')
plt.show
```

# 선택한 지역의 최저기온과 최고기온 분석하기


1.   선택한 지역 데이터 준비하기
*   선택한 이유

2.   다운로드 설정
* 스샷 (검색조건)

3. 데이터 형태
* 10줄 정도(노트패드) 스샷

4. 선택한 지역의 최저 기온과 최고 기온
* 기상관측 아래 서울의 기온이 ~~
* 기상관측 아래 서울의 기온이 ~~
* 최고 기온 전체 그래프 그리기

```python
import csv
min_temp = 999
min_date = ''
f = open('/content/Deagwanryeong.csv', 'r', encoding='cp949')
data = csv.reader(f)
header = next(data)
print(header)


for row in data:
  if row[-2] == '':
    row[-2] = 999
  row[-2] = float(row[-2])

  if min_temp > row[-2]:
    min_date = row[0]
    min_temp = row[-2]

f.close()

print(f'대관령의 기온이 가장 낮았던 날은 {min_date}이고 최저 기온은 {min_temp}입니다.')
```
