step2. 데이터 중 최고 기온을 실수로 변환하여 한 행씩 출력하기

step3. 최고기온과 최고기온이었던 날짜 찾기

```python
import csv
max_temp = -999
# 최고 기온 값을 저장할 변수
# 명확한 결과를 위해 극적인 값으로 셋팅을 함

max_date = ''
# 최고 기온이 가장 높았던 날짜를 저장하는 변수

f = open('/content/seoul.csv', 'r', encoding='cp949')

data = csv.reader(f)
header = next(data)

for row in data:
  # 최고 기온 데이터가 '' (결측치) 인지 확인
  if row[-1] == '':
    row[-1] = -999
    # -999를 넣어서 빈 문자열('')이 있던 자리라고 표시한다.'
  row[-1] = float(row[-1]) # 기온 데이터 실수형으로 변환

  # 최고기온 정보를 저장
  
  if max_temp < row[-1]:
    max_date = row[0]
    max_temp = row[-1]

f.close()

print(f'기상 관측 이래 서울 최고 기온이 가장 높았던 날은 {max_date} 최고기온은 {max_temp}입니다.' )
```

step4. 최저기온과 최저기온이었던 날짜 찾기

```python
import csv
min_temp = 999
# 최저기온 값을 저장할 변수
# 명확한 결과를 위해 극적인 값으로 셋팅을 함

min_date = ''
# 최저기온이 가장 높았던 날짜를 저장하는 변수

f = open('/content/seoul.csv', 'r', encoding='cp949')

data = csv.reader(f)
header = next(data)

for row in data:
  # 최저기온 데이터가 '' (결측치) 인지 확인 // row[-2]: 최저기온 // row[0]: 날짜
  if row[-2] == '':
    row[-2] = 999
    # 999를 넣어서 빈 문자열('')이 있던 자리라고 표시한다.'
  row[-2] = float(row[-2]) # 기온 데이터 실수형으로 변환
  # 실수형으로 바꿨기 때문에 min_temp와 비교가 가능하다.

  # 최저기온 정보를 저장
  
  if min_temp > row[-2]:
    min_date = row[0]
    min_temp = row[-2]

# min_temp가 999도이고 999가 > 7.9도가 참이면 
# 1909-10-01이 비워져있던 min_date에 들어간다.
# 7.9도 이던 row[-2]가 999도 이던 min_temp로 들어간다,

# min_temp가 7.9도이고 7.9가 > 13.1도 보다 작기때문에 False가 된다. 
# False이므로 if문을 수행하지 않는다.

# min_temp가 7.9도이고 7.9가 > 6.3도가 참이면 
# 1909-10-07이 비워져있던 min_date에 들어간다.
# 6.3도 이던 row[-2]가 7.9도 이던 min_temp로 들어간다,


f.close()

print(f'기상 관측 이래 서울 최고 기온이 가장 높았던 날은 {min_date} 최고기온은 {min_temp}입니다.' )
```
