꺾은 선 그래프로 표현하기

```python
import csv
import matplotlib.pyplot as plt


f = open('/content/gender.csv', 'r', encoding='cp949')
data = csv.reader(f)
header = next(data)

name = input('지역을 입력하시오.   :')

m = []
f = []



for row in data:
  if name in row[0]:
    for i in row[3:104]:
      m.append(int(i.replace(',', '')))
    for j in row[106:]:
      f.append(int(j.replace(',', '')))
    break

f.reverse
# 0세부터 인구수 나올 수 있도록 데이터 뒤집기

print(m)
print(f)

plt.figure(dpi=200)
plt.style.use('ggplot')

# 꺾은 선 그래프로 표현하기

plt.plot(m, label='남성')
plt.plot(f, label='여성')
plt.legend()
plt.show()
```

연령별로 남성 인구수에서 여성 인구수를 뺀 차이의 막대그래프 그리기


```python
import csv
import matplotlib.pyplot as plt


f = open('/content/gender.csv', 'r', encoding='cp949')
data = csv.reader(f)
header = next(data)

name = input('지역을 입력하시오.   :')

result = [] # 남성 인구 - 여성 인구 값을 담을 리스트 변수 선언

for row in data:
  if name in row[0]:
    for i in range(3,104):
      result.append(int(row[i].replace(',', '')) - int(row[i+103].replace(',', '')))
    break

plt.figure(dpi=200)
plt.style.use('ggplot')
plt.title(name + '지역의 남녀 성별 인구 차이 분포')
plt.bar(range(101), result)
plt.show()
```

원의 크기는 인구수의 제곱근 추가


```python
import csv
import matplotlib.pyplot as plt
import math

f = open('/content/gender.csv', 'r', encoding='cp949')
data = csv.reader(f)
header = next(data)

name = input('지역을 입력하시오.   :')

m = []
f = []
size = [] # 인구수 사이즈 리스트 변수 선언


for row in data:
  if name in row[0]:
    for i in row[3:104]:
      m.append(int(i.replace(',', '')))
      size.append(math.sqrt(int(i.replace(',', ''))))
    for j in row[106:]:
      f.append(int(j.replace(',', '')))
    break

# 인구수와 math.sqrt(인구수) 차이 확인하기
print(m)
print(size)
# 스캐터 그래프로 표현하기
plt.figure(dpi=200)
plt.scatter(m, f, s=size, c=range(101), cmap='jet', alpha=0.7)
# x축은 남자, y축은 여자
plt.colorbar()
plt.plot(range(max(m)), range(max(m)), 'g')
plt.show()
```

```python
#한준이 코드
import csv
import matplotlib.pyplot as plt

f = open("gender.csv", encoding="cp949")
data = csv.reader(f)
result = []
location = ["서울특별시","부산광역시", "대구광역시", "인천광역시", "광주광역시", "대전광역시", "울산광역시", "세종"]

#name=input("찾고 싶은 지역의 이름을 알려주세요 :")
for name in location:
    globals()[f"result_{name}"] = []
    for row in data: 
        if name in row[0]:
            for i in range(3, 104):
                globals()[f"result_{name}"].append(int(row[i].replace(",","")) - int(row[i+103].replace(",","")))
            break

fig = plt.figure(dpi=100, figsize=(10,16))        #### plt.figure(dpi=100, figusize=(10, 16))을 fig라는 객체에 담아서 틀을 만들어줌
plt.rc('font', family = 'NanumBarunGothic')
plt.style.use('ggplot')

for idx, name in enumerate(location):
    globals()[f"plot_{name}"] = fig.add_subplot(4, 2 ,idx+1)  #### 아까 만든 fig 틀에서 .add_subplot(행, 열, 인덱스 넘버(위치를 지정해줌//그래프는 인덱스가 1부터 시작함!!))를 사용하여 그것을 객체에 담음 - 8개 이면 8번 작성해줘도 됨 // 하지만 globals()를 이용하여 반복을 한 것임
    globals()[f"plot_{name}"].bar(range(101),globals()[f"result_{name}"])
    globals()[f"plot_{name}"].set_title(f"{name}")
    globals()[f"plot_{name}"].set_ylim(-10000, 4000)
plt.legend()
plt.show()
```
