지역명을 입력받아서 연령별 인구수 데이터 시각화하기

```python
import csv
import matplotlib.pyplot as plt


f = open('/content/age.csv', 'r', encoding='cp949')
data = csv.reader(f)
header = next(data)

result = []

# 지역명을 입력받는 변수 선언

name = input('인구 구조가 알고 싶은 지역의 이름(읍면동 단위)를 입력해주세요. : ')

for row in data:
  if name in row[0]:
    for i in row[3:]:
      new_i = i.replace(',', '')
      result.append(int(new_i)) 

# 인구구조 시각화

plt.figure(dpi=200)
plt.style.use('ggplot')  # 격자무늬 스타일로 지정  classic, dark_background, ggplot
plt.title(name + ' 지역의 인구 구조')
plt.plot(result)
plt.show()
```

2. 항아리 모양 그래프 그리기
* 성별 데이터 저장하기

```python
# 남성 연령별 인구수 - 3 ~ 103 // 여성 연령별 인구수 - -101 ~ -1

import csv
import matplotlib.pyplot as plt


f = open('/content/gender.csv', 'r', encoding='cp949')
data = csv.reader(f)
header = next(data)

m = []
f = []

for row in data:
  if '신도림' in row[0]:
    for i in range(0, 101):
      m.append(int(row[i + 3]))
      f.append(int(row[-(i+1)])) # -1 값부터 담는다. -> 100세 이상부터 값이 담긴다.



f.reverse() 
# 0세부터 인구수 나올 수 있도록 데이터 뒤집기

print(m)
print(f)
```

```python

import csv
import matplotlib.pyplot as plt


f = open('/content/gender.csv', 'r', encoding='cp949')
data = csv.reader(f)
header = next(data)

m = []
f = []

for row in data:
  if '신도림' in row[0]:
    for i in range(0, 101):
      m.append(-int(row[i + 3])) # 값들이 음수로 들어간다.
      f.append(int(row[-(i+1)])) # -1 값부터 담는다. -> 100세 이상부터 값이 담긴다.



f.reverse() 
# 0세부터 인구수 나올 수 있도록 데이터 뒤집기

print(m)
print(f)

plt.figure(dpi=200)
plt.style.use('ggplot')  # 격자무늬 스타일로 지정  classic, dark_background, ggplot
plt.rcParams['axes.unicode_minus'] = False # x축의 마이너스 ticks를 알려준다. 
plt.barh(range(101), m, label='남')
plt.barh(range(101), f, label='여')
plt.legend()
plt.title('신도림 지역의 남녀 성비 구분')
plt.show()

```
