statisctics 모듈을 사용해서 나이 변수의 평균과 표준편차 구하기



```python
import csv

with open('/content/Demographic_Statistics_By_Zip_Code.csv') as infile:
  data = list(csv.reader(infile))
```

```python
# 첫 번째 레코드에 있는 'COUNT PARTICIPANTS' 데이터 인덱스 값 추출 data[0] 컬럼 값
# data[0]은 데이터의 0번째 인덱스 즉, 컬럼들을 가리킨다.

COUNT_PARTICIPANTSindex = data[0].index('COUNT PARTICIPANTS')
COUNT_PARTICIPANTSindex
# 인덱스를 찾아냈음
```

```python
# 'COUNT PARTICIPANTS'의 데이터를 추출함
# data[1:] 하는 이유는 0번째 인덱스는 컬럼명이 있으니까
# row[COUNT_PARTICIPANTSindex]는 특정 컬럼 값만 추출한다.
# int(row[COUNT_PARTICIPANTSindex])를 해서 정수형으로 만들어줘야 계산할 수 있음(명시적으로 int()를 사용하라!)
# 숫자인데 ' ' 있으면 문자형인 것임

COUNTPARTICIPANTS = [int(row[COUNT_PARTICIPANTSindex]) for row in data[1:]] # 리스트 내포
COUNTPARTICIPANTS
```

나이의 평균과 표준편차를 구하라

```python
import statistics
print(f'평균 나이: {statistics.mean(COUNTPARTICIPANTS)}, 표준편차: {statistics.stdev(COUNTPARTICIPANTS)}')
```

```python
wn = nltk.corpus.wordnet
# 코퍼스 리더(reader)

# synset : 유의어 묶음으로 단어 - 품사 - 순번으로 구성
wn.synsets('cat')
```

```python
wn.synset('cat.n.01').hypernyms()
# hypernyms()는 상위어를 의미
```

```python
wn.synset('cat.n.01').hyponyms()
# hyponyms()는 하위어를 의미
```

Wordnet을 사용해서 synset 간 의미론적 유사도 계산하기(0-1 사이의 실수)</br>
0이면 두 단어 서로 관계 없음 </br>
1이면 완전한 유의어

```python
# 고양이, 링스 : 시라소니 유사도 계산하기
x = wn.synset('cat.n.01') # 고양이
y = wn.synset('lynx.n.01') # 시라소니

# x, y의 유사도 계산하기
x.path_similarity(y)
```

```python
# 집고양이, 길고양이의 유사도 계산하기
x = wn.synset('domestic_cat.n.01')
y = wn.synset('wildcat.n.01')

# x, y의 유사도 계산하기
x.path_similarity(y)
```

```python
# 강아지, 고양이의 유사도 계산하기
x = wn.synset('dog.n.01')
y = wn.synset('cat.n.01')

# x, y의 유사도 계산하기
x.path_similarity(y)
```

####정규화: 추가적으로 데이터를 처리하려고 자연어로 된 텍스트를 준비하는 과정

###정규화 단계
1.   토큰화: 텍스트를 단어로 쪼갬
2.   단어의 대, 소문자 통일: 전부 다 대문자 or 소문자
3.   불용어 제거: The와 같은 분석에 큰 의미가 없는 단어
4.   형태소 분석: 단어를 형태소로 변환
5.   원형 추출
6.   그 외 품사 태깅 POS tagging(Part Of Speech tagging) 중요함



```python
# 1. 토큰화 (텍스트 -> 단어를 쪼갠다.)
from nltk.tokenize import WordPunctTokenizer
word_punct = WordPunctTokenizer()

text = "}Help! :))) :[ ..... :D{"

# Tokenizer는 모든 구두점(punctuation, 문장부호)을 기준으로 분리한다. 
# -> 이모티콘을 이용한 감성 분석 등 문장구조 깊이 분석할 때 사용

word_punct.tokenize(text)
```

```python
import nltk
nltk.download('punkt')
```

```python
# 단어 토큰나이저를 사용하기 
# 단어를 중심으로 자름
# 문장부호와 상관없이 싹 다 나눈다.
# 단어만 추출할 때 사용하기 좋다.
nltk.word_tokenize(text)
```

```python
# 포터 형태소 분석기(보수적)

pstemmer = nltk.PorterStemmer()
pstemmer.stem('wonderful')
```

```python
# 랭커스터 형태소 분석기 (적극적) - 더 많은 동음이의어 형태소 생산

lstemmer = nltk.LancasterStemmer()
lstemmer.stem('wonderful')
```

```python
# 5. 원형 추출
lemmatizer = nltk.WordNetLemmatizer()
lemmatizer.lemmatize('wonderful')
```

```python
# 그 외의 품사 태깅
nltk.download('averaged_perceptron_tagger')
nltk.pos_tag(["beautiful", "world"]) # JJ	adjective(형용사), NN	noun(명사)
# https://happygrammer.github.io/nlp/postag-set/

```

```python
# www.networksciencelab.com 사이트에서 하이퍼 링크 정보 모두 가져오기

from bs4 import BeautifulSoup
from urllib.request import urlopen

# open()함수를 통해서 파일을 열때, 꼭 close()를 써줘야 한다.
# 하지만 with문을 이용해서 파일을 열거나 쓰고나면 close()를 하지 않아도 된다.
# with ~ as 문장을 실행하고 나면 자동으로 close()를 호출하기 때문이다.

with urlopen('http://www.networksciencelab.com/') as doc:
  soup = BeautifulSoup(doc)

# 하이퍼링크만 추출하기 (list comprehension으로 한 줄로 끝냄)
links = [(link.string, link['href']) for link in soup.find_all('a') if link.has_attr('href')]

# 책 제목만 추출하기
html_text = [i[0] for i in links]
# i의 0번째 인덱스를 추출하라 / links에 있는 값들을 i에 넣으면서
# 지금 links에는 string과 href가 있지만 하이퍼텍스트는 필요 없으므로 string만 추출하는 과정이다.

# None 데이터 제거 -> 리스트에서 텍스트로 변환
html_text = ' '.join((html_text[2:]))
# 공백을 이용해서 조인을 하라 / html_text의 2번째 인덱스부터
# ' '.join(' ')을 이용하면 리스트를 텍스트로 만든다.

# 텍스트 분석을 하기 위해서는 텍스트이어야 하기 때문에 join을 사용해서 하나의 텍스트로 만들어줬다.
html_text
```

```python
# 하나의 텍스트 -> index.html 파일로 저장하기
html_file = open('index.html', 'w')
html_file.write(html_text)
html_file.close()
```

Index.html 파일에서 (불용어 제외) 가장 많이 등장한 단어 원형 찾아보기

```python
from bs4 import BeautifulSoup # 추출하기
from collections import Counter # 추출된 것 중 숫자 세기
from nltk.corpus import stopwords # 불용어를 삭제하기 위해서
from nltk import LancasterStemmer # 원형데이터 가져오기

# 형태소 분류기 생성하기
ls = nltk.LancasterStemmer()

# 파일을 읽고 soup을 만든다.
with open('/content/index.html') as infile:
  soup = BeautifulSoup(infile)
# html을 읽기 위해서는 BeautifulSoup을 이용해서 읽음

```

```python
# 1). 텍스트를 추출하기 -> 토큰화하기
# 변수명 잘 짓기! - 책에서 나온 스타일대로...
# 무조건 처음은 소문자, 이어질 부분의 첫글자는 대문자 (camel형식)
# countParticipantsIndex
words = nltk.word_tokenize(soup.text)
words
```

```python
# 2) 단어를 소문자로 변환
words = [w.lower() for w in words]
words
```

```python
import nltk
nltk.download('stopwords')
```

```python
# 3) 불용어를 제거하고 단어의 형태소를 추출하라
# 3-1) 불용어 제거 not in stopwords.words('english')
# 3-2) 특수기호 isalnum()
words = [ls.stem(w)for w in words if w not in stopwords.words('english') and w.isalnum()]
# isalnum : 문자열이 알파벳[a-z A-Z]과 숫자[0-9]로만 구성되었는가?
words
```

```python
# 4) 가장 빈번하게 등장하는 단어 10개 추출
freqs = Counter(words)
print(freqs.most_common(10))
```
